# Distribution of stats of NBA

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('players_stats_by_season_full_details.csv')
```

```python
df.head()
```

```python
db = df[df['League'] == 'NBA']
```

### We have everything set, now we shall start the analysis, getting the mean of each numerical stats, which will be used on other parts of the project.

```python
dic_mean = {'GP': db['GP'].mean(),'MIN': db['MIN'].mean(), 'FGM': db['FGM'].mean(), 'REB': db['REB'].mean(), 'FGA': db['FGA'].mean(),
           '3PM': db['3PM'].mean(),'3PA': db['3PA'].mean(),'FTM': db['FTM'].mean(),'FTA': db['FTA'].mean(),'TOV': db['TOV'].mean(),'PF': db['PF'].mean(),'ORB': db['ORB'].mean(),
           'DRB': db['DRB'].mean(),'AST': db['AST'].mean(),'STL': db['STL'].mean(),'BLK': db['BLK'].mean(),'PTS': db['PTS'].mean(),'height_cm': db['height_cm'].mean(),
           'weight_kg': db['weight_kg'].mean()}
```

#### Let's check the mean for height and weight, and later analyze the distribution of them.

```python
[dic_mean['height_cm'], dic_mean['weight_kg']]
```

```python
figure, plot = plt.subplots(1,2)
sns.displot(data=db, x = 'height_cm', ax = plot[0], color='green')
sns.displot(data=db, x = 'weight_kg', ax = plot[1], color= 'red')
plot.show()
```
